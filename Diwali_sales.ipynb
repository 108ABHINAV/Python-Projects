```python
#import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#droping irrevelent columns

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#removing null values
df.dropna(inplace=True)
```

# Exploratory Data Analysis

# 1- Gender

```python

ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

# 2- Age

```python
df.columns
```

```python
ax= sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount v age group

sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From above graphs most buyers are between age group between 26-35 yrs females

## 3- State

```python
df.columns
```

```python
#total numbers of orders from top 10 states

sales_st=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_st)
```

```python
#total sales from top 10 states

sales_st=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_st)
```

from the above graphs we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharastra, karnataka respectively

## 4- Martial Status

```python
ax= sns.countplot(data=df,x='Marital_Status',hue='Gender')

sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ma=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(7,4)})
sns.barplot(data=sales_ma,x='Marital_Status',y='Amount')
```

from above graphs we can see that most of the buyers are married(women) and they have high purchasing power

# 5- Occupation

```python
ax= sns.countplot(data=df,x='Occupation')

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ma=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_ma,x='Occupation',y='Amount')
```

from above graphs we can see that most of the buyers are working in IT, Healthcare and aviation sector

```python
df.columns
```

# 6- Product Category

```python
ax= sns.countplot(data=df,x='Product_Category')

sns.set(rc={'figure.figsize':(30,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ma=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data=sales_ma,x='Product_Category',y='Amount')
```

from above graphs we can see that most of the sold products are from food, footwear and electronics category

```python
#top 10 sold product id's
sales_ma=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data=sales_ma,x='Product_ID',y='Orders')
```

# Conclusion:


Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
